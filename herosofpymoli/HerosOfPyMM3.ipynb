* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#purchase analysis frame
#create group by gender
grouped_gender = purchase_data.groupby(['Gender'])
#count number of purchases
purchase_count = grouped_gender["Purchase ID"].count()
#putting price into gender grouping
price_sum = grouped_gender["Price"].sum()
#getting average of price by group
price_avg = purchase_data['Price'].mean()
#calculating average price by group
avg_price = (price_sum / purchase_count)

gender_totals = purchase_data["Gender"].value_counts()
Average_Purchase_Total =  price_sum/gender_totals


#create frame
purch_table = pd.DataFrame({"Purchase Count":purchase_count, "Average Purchase Price":avg_price, 
                            "Total Purchase Value":price_sum, "Avg Total Purchase per Person":Average_Purchase_Total})
#assign formatting
purch_table["Average Purchase Price"] = purch_table["Average Purchase Price"].map("${:.2f}".format) 
purch_table["Total Purchase Value"] = purch_table["Total Purchase Value"].map("${:.2f}".format)
purch_table["Avg Total Purchase per Person"] = purch_table["Avg Total Purchase per Person"].map("${:.2f}".format)
purch_table
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#create the bin
age_bin = [0, 9, 14, 19, 24, 29, 34, 39, 130]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39","40+"]
purchase_data["Age_Ranges"] = pd.cut(purchase_data["Age"], age_bin, labels = group_names)
#calculate unique players - give a total
total_players = len(purchase_data["SN"].unique())
#calculate unique players using drop - removes from the list
purch_without_duplicates = purchase_data.drop_duplicates(["SN"]) 
#create group from bin
ttl_age = purch_without_duplicates.groupby("Age_Ranges")
#put counts into group
total_age = (ttl_age["SN"].count())
#put percents into group and format
pct_age = ((ttl_age["SN"].count()/total_players)).map("{:.2%}".format)

#make the frame
purch_age = pd.DataFrame({"Age Count":total_age, "Age Percent":pct_age})

purch_age                       
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#EDIT HERE
#Use Bin from the purchase_data data frame by age from above
purchase_data_age= purchase_data.groupby("Age_Ranges")
#count purchases
purc_count_age = purchase_data_age["Item ID"].count()
#determine average price per person (sum all prices / purchase count)
avg_purc_pr_age = (purchase_data_age["Price"].sum()/purc_count_age).map("${:.2f}".format)
#total price broken by age ranges
ttl_purc_price_age = (purchase_data_age["Price"].sum()).map("${:.2f}".format)
#average by age range = total price divided by count f age 
avg_purc_ttl_age = (purchase_data_age["Price"].sum()/total_age).map("${:.2f}".format)
# make the frame
summary_purchase_age = pd.DataFrame({"Purchase Count": purc_count_age, "Average Purchase Price": avg_purc_pr_age,
                                       "Total Purchase Value": ttl_purc_price_age, "Avg Total Purchase per Person": avg_purc_ttl_age})
summary_purchase_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Create a summary data frame to hold the results Purchase ID
purchase_data_age= purchase_data.groupby("SN")

purc_count_item = purchase_data_age["Item ID"].count()

#determine average price per person (sum all prices / purchase count)
avg_purc_pr_age = (purchase_data_age["Price"].sum()/purc_count_item).map("${:.2f}".format)
total_purch_val = purchase_data_age["Price"].sum()

#sort in decending order

#put in frame and format
purch_table3 = pd.DataFrame({"Purchase Count":purc_count_item, "Average Purchase Price":avg_purc_pr_age, 
                            "Total Purchase Value":total_purch_val}).sort_values(by='Total Purchase Value', ascending=False)

#assign formatting
purch_table3["Total Purchase Value"] = purch_table3["Total Purchase Value"].map("${:.2f}".format) 


purch_table3.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
pop_items = pd.DataFrame({'Item ID' : purchase_data['Item ID'], 'Item Name' : purchase_data['Item Name'], 
                             'Item Price' : purchase_data['Price']})
pop_items_group = pop_items.groupby(['Item ID', 'Item Name'])

pop_items_count = pop_items_group.count()['Item Price']
pop_items_count

pop_tot_value = pop_items_group.sum()['Item Price']
pop_tot_value

#Put calculations into a summary dataframe
pop_summary_df = pd.DataFrame({'Purchase Count' : pop_items_count, 'Total Purchase Value': pop_tot_value})
#pop_summary_df.sort_values(by=['Purchase Count'], ascending=False)

pop_merge_df = pd.merge(pop_items_df, pop_summary_df, on=['Item ID'])
pop_merge_df = pop_merge_df.drop_duplicates()
pop_merge_df = pop_merge_df[['Item ID', 'Item Name', 'Purchase Count', 'Item Price', 'Total Purchase Value']]
most_popular_items = pop_merge_df.sort_values(by=['Purchase Count'], ascending=False).head()

#reset the index so that we can format total purchase value and keep data sorted with the new index
new_pop_index = most_popular_items.reset_index(drop=True)
new_pop_index['Total Purchase Value'] = new_pop_index['Total Purchase Value'].astype(float).map("${:,.2f}".format)
new_pop_index.head()
```
